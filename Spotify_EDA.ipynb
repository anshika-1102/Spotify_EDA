```python
#Importing required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the Spotify dataset
df_tracks = pd.read_csv('SpotifyFeatures.csv')
df_tracks.head()
```

```python
#Check for null values
pd.isnull(df_tracks).sum()
```

```python
#Information about dataset (total number of rows and columns , data type)
df_tracks.info()
```

```python
# 10 least popular songs in Spotify playlist
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
#descriptive statistics
df_tracks.describe().transpose()
```

```python
# 10 most popular songs with popularity>90
most_popular = df_tracks.query('popularity>50', inplace = False).sort_values('popularity', ascending = True)
most_popular[:10]
```

```python
#check artist at 18th row
df_tracks[["artists"]].iloc[18]
```

```python
#convert duration in milliseconds to seconds
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace= True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
#correlation map between variables
corr_df = df_tracks.drop(["key","mode","explicit","artists","name","id","release_date","year"],axis=1).corr(method = "pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidth=1, linecolor="Black")
heatmap.set_title("Correlation heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#Regression line between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y="loudness", x="energy", color="c").set(title= "Loudness vs Energy correlation")
```

```python
#Regression line between popularity and acousticness
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y="popularity", x="acousticness", color="b").set(title= "Popularity vs Acousticness correlation")
```

```python
#distribution plot
sns.displot(df_tracks["year"], discrete=True, aspect=2, height=5, kind="hist").set(title = "Number of songs per year")
```

```python
#bar plot for duration of songs
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = df_tracks['year'], y = total_dr, ax=ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation=90)
```

```python
#Line plot to find average duration of songs over the years
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = df_tracks["year"], y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre = pd.read_csv("data_genre.csv")
```

```python
df_genre.head()
```

```python
#duration of songs with different genres - bar plot
plt.title("Duration of songs with different genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
#top 3 genres by popularity
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending=False).head(90)
sns.barplot(y='genre', x='popularity', data=famous).set(title="Top 5 genres by popularity")
```
